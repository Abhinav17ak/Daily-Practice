# Banking Syststem

```python
from abc import ABC , abstractmethod

class Account(ABC):
    def __init__(self , acc_no , pin , balance = 0):
        self.acc_no = acc_no
        self.__pin = pin
        self.__balance = balance
        
        
# Pin Authentication
    def __verify_pin(self , pin):
        return self.__pin  == pin
    
# Protectect Balance Access
    def _get_balance(self) :
        return self.__balance
    
    def _set_balance(self , amount) :
        self.__balance = amount
        
# Getter Functions
    def get_balance(self , pin):
        if self.__verify_pin(pin):
            return self.__balance
        raise PermissionError("Invalid PIn")
    
# Common metjods of each accounts
    def deposit(self , amount ,  pin) :
        if not self.__verify_pin(pin):
            raise PermissionError("Invalid PIn")
        
        if amount <= 0 :
            raise ValueError("Deposit Must be positive")
        
        self.__balance += amount
        print(f"Deposited Amount :- {amount} , and Balanve :- {self.__balance}")
        
    @abstractmethod
    def withdraw(self , amount , pin):
        pass
    
    @abstractmethod
    def calculate_interest(self):
        pass
    
    
class SavingsAccount(Account):
    def __init__(self, acc_no, pin, balance=0):
        super().__init__(acc_no, pin, balance)
            
    # Method Over riding
    
    def withdraw(self, amount, pin):
        if amount <= 0 :
            raise ValueError("Invalid Withdrawl Amount")
        
        if self.get_balance(pin) <amount :
            raise ValueError("insufficient Balance")
        
        self._set_balance(self._get_balance() - amount)
        print(f"withdraw Amount {amount} , Balance = {self._get_balance()}")
        
        
    def calculate_interest(self):
        interest = self._get_balance() * 0.04
        return interest
    
    
class CurrentAccount(Account) :
    def __init__(self, acc_no, pin, balance=0 , overdraft_limit = 5000):
        super().__init__(acc_no, pin, balance) 
        self.overdraft_limit = overdraft_limit
        
    # method overriding
    
    def withdraw(self, amount, pin):
        if amount <= 0:
            raise ValueError("Invalid Withdrawl Amount")
        
        available = self._get_balance() + self.overdraft_limit
        if amount > available :
            raise ValueError("Overdraft limit exceeded")
        
        self._set_balance(self._get_balance() - amount)
        print(f"withdraw {amount} . Balance :- {self._get_balance()}")
    
    def calculate_interest(self):
        return 0
    
    
    
# Driver Code

s1 = SavingsAccount("SAV-001" , 1234 , 10000)
c1 = CurrentAccount("Curr-201" , 9999, 5000)

print("\n--- Savings Account ---")
s1.deposit(2000, 1234)
s1.withdraw(3000, 1234)
print("Interest:", s1.calculate_interest())
            
print("\n--- Current Account ---")
c1.withdraw(7000, 9999)  
print("Balance:", c1.get_balance(9999))
            
        
        
        
        
        
        
        
```
