Constructor & Inheritance

Create a class Employee and subclasses Manager and Developer. Initialize attributes using __init__() and demonstrate constructor chaining using super().

Hint: Show how parent class attributes are accessed in child class

```python
# base Class
class Employee:
    def __init__(self , name , salary):
        self.name  = name 
        self.salary = salary
        print(f"Employee {self.name} is initialized with salary {self.salary}")
        
# subclass Manager

class Manager(Employee):
    def __init__(self , name , salary , department):
        super().__init__(name ,salary)# constructor chaning
        self.department = department
        print(f"Manager {self.name} of  {self.department} department initilized ")
        
# subclass Developer
class Developer(Employee):
    def __init__(self , name , salary , language):
        super().__init__(name , salary)
        self.language = language
        print(f"Developer {self.name} who codes in {self.language}")
        
m = Manager("Abhinav" , 900000 , "HR")
d = Manager("Abhinav2" , 700000 , "Python")
```
