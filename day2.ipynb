Reverse Words in a Sentence

```python
s1 = "Python is easy"
list_s1 = s1.split(" ")
res = ""
for i in list_s1:
    res = res + i[::-1] + " "
print(res)
```

Find the second largest number in a list without using sort()

```python
l1 = [10, 5, 20, 8]
for i in range(1 ,len(l1)):
    if l1[i-1] > l1[i]:
        l1[i-1]  , l1[i] = l1[i] , l1[i-1]
print(l1[-2])
```

Q7. Sum of Digits

```python
n = 1234
sum = 0
while n > 0 :
    ld = n % 10
    sum += ld
    n //= 10
print(sum)
```

Q9. Check Anagram

```python
s1 = "listen"
s2 = "silent"

if len(s1) != len(s2):
    print("Not Anagram")
else :
    word_s1 = s1.split().sort()
    word_s2 = s1.split().sort()
    if word_s1 == word_s2:
        print("Anagram")
```

Q13. Fibonacci Series (N terms)

```python
a = 0
b = 1
n = int(input("enter n"))
if n == 1:
    print( a)
elif n == 2 :
    print( a , b)
elif n > 0:
    print(a , b , end= " ")
    for i in range(2 , n):
        c = a+b 
        a = b
        b = c
        print(c , end = " ")
        
```

Q15. Check Prime Number

```python
n = 25
if n < 2 :
    print("Enter no 2 or more than 2")
else:
    for i in range( 2 , n):
        if n % i == 0:
            print("NOT Prime")
            break
    else:
        print("Prime")
```
