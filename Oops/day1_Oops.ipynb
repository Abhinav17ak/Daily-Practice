Q1. Encapsulation + Validation

Design a BankAccount class

Private balance

deposit() and withdraw() methods

Withdraw should fail if balance is insufficient

ðŸ‘‰ Tests:
Encapsulation, data hiding, validation logic

```python
class BankAccount:
    def __init__(self ,name , balance = 0):
        self.name = name 
        self.__balance = balance # private variable
    
    def deposit(self , amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited Amount {amount}")
            
        else:
            print("Invalid Deposit amount")
    
    def withdraw(self , amount) :
        if amount <= self.__balance :
            self.__balance -= amount
            print(f"withdraw Amount {amount}")
        else : 
            print("Invalid Amount")
            
    def get_balance(self):
        return self.__balance
    
    
# Usage

acc = BankAccount("Abhinav" ,10000 )
acc.deposit(12)
acc.get_balance()
acc.withdraw(12)
acc.get_balance()
            
            
        
```

Q3. Inheritance + Method Overriding

Create:

Employee class â†’ calculate_salary()

Manager class â†’ adds bonus

ðŸ‘‰ Tests:
Inheritance, method overriding, super()

```python
class Employee:
    def __init__(self , name  , salary ):
        self.name = name 
        self.salary = salary
    
    def calculate_salary(self):
        return self.salary
    

class Manager(Employee) :
    def __init__(self , name  , salary , bonous):
        super().__init__(name , salary)
        self.bonous = bonous
        
    def calculate_salary(self):
        return  self.salary + self.bonous
    
    
emp = Employee("Rohit" , 30000)
mgr = Manager("Anjali" , 500000 , 1000)

print("Employee salary" , emp.calculate_salary())
print("manager salary" , mgr.calculate_salary())
        
```

Q5. Abstract Class (Real-World)

Create an abstract class Payment:

pay(amount) (abstract method)

Subclasses:

UPI

CreditCard

Cash

ðŸ‘‰ Tests:
Abstraction, interface-like behavior

```python
from abc import ABC , abstractmethod

class Payment(ABC):
    @abstractmethod
    def pay(self , amount):
        pass

class UPI(Payment):
    def pay(self , amount):
        print(f"Paid {amount} using UPI")
        
class CreditCard(Payment):
    def pay(self, amount):
        print(f"Paid â‚¹{amount} using Credit Card")


class Cash(Payment):
    def pay(self, amount):
        print(f"Paid â‚¹{amount} in Cash")

        
payments = [UPI() ,CreditCard() , Cash() ]

for p in payments :
    p.pay(1000)
```

```python
class Logger:
    def log(self):
        print("Logging data")

class DataBase:
    def log(self) :
        print("saving data to the Data Base")

class App(Logger, DataBase):
    def log(self):
        Logger.log(self)
        DataBase.log(self)
        
app = App()
app.log()
```
