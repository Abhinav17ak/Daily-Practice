ðŸ”¹ Question 1: Student Management System
Problem

Design a system to manage students:

Store multiple students

Each student has marks of multiple subjects

Calculate average marks

Different student types calculate grade differently

```python
from abc import ABC ,abstractmethod

class Students(ABC):
    def __init__(self , name , marks):
        self.name = name 
        self.__marks = marks
        
    def get_average(self):
        return sum(self.__marks.values())/len(self.__marks)
    
    @abstractmethod
    def calculate_grade(self):
        pass
    
class RegularStudents(Students):
    def calculate_grade(self):
        avg = self.get_average()
        return "A" if avg >= 80 else "B"
 
class ScholarShipStudents(Students):
    def calculate_grade(self):
        avg = self.get_average()
        return "A+" if avg >= 75 else "A"   
    
    
s1 = RegularStudents('Abhinav' ,{"Math": 80, "DS": 85} )
s2 = ScholarShipStudents("Ravi", {"Math": 78, "DS": 82})
print(s1.name, s1.calculate_grade())
print(s2.name, s2.calculate_grade())
```

âœ… Smart Parking System â€” Full Implementation Using All OOP Concepts
âœ” Abstraction â†’ Abstract class Vehicle
âœ” Inheritance â†’ Bike, Car, Truck
âœ” Polymorphism â†’ Each vehicle overrides calculate_fee()
âœ” Encapsulation â†’ Private variables (__slots, __in_time, __out_time)
âœ” Composition â†’ ParkingLot manages multiple vehicles
âœ” Exception Handling

```python
from abc import ABC , abstractmethod
import time
class Vehicle(ABC):
    def __init__(self , number):
        self.number = number
        self.__in_time = None
        self.__out_time = None
    
    def entry(self):
        self.__in_time = time.time()
        
    def exit(self):
        self.__out_time = time.time()
        
    # Encapsulated Getters 
    def get_parking_duration(self):
        if self.__out_time is None or self.__in_time is None :
            return 0
        return (self.__out_time - self.__in_time) / 60
    
    @abstractmethod
    def calculate_fee(self , minutes):
        pass
    

class Bike(Vehicle):
    def calculate_fee(self, minutes):
        return 10 + 0.5 * minutes
    
    
class Car(Vehicle):
    def calculate_fee(self, minutes):
        return 20 + 1.0 * minutes

class Truck(Vehicle):
    def calculate_fee(self, minutes):
        return 50 + 2.0 * minutes
     

# Parking Lot Class Encapsulation

class ParkingLot:
    def __init__(self , total_slots):
        self.__total_slots = total_slots
        self.__available_slots = total_slots
        self.active_vehicles = {}
        
    # Encapsulated slot Count
    def get_available_slots(self):
        return self.__available_slots
    
    def vehicle_entry(self , vehicle):
        if self.__available_slots <= 0:
            print("No Slots Available")
            return False
        
        vehicle.entry()
        self.active_vehicles[vehicle.number] = vehicle
        self.__available_slots -= 1
        print(f"vehicle {vehicle.number} entered. \n Slots Left :{self.__available_slots} \n **************************")
        return True
    
    def vehicle_exit(self , number):
        if number not in self.active_vehicles :
            print("Vehicle Not Found")
            return False
        vehicle = self.active_vehicles[number]
        vehicle.exit()
        
        # Release Slots
        
        self.__available_slots += 1
        minutes = vehicle.get_parking_duration()
        fee = vehicle.calculate_fee(minutes)
        
        print("\n========== BILL ==========")
        print("Vehicle No:", number)
        print(f"Parking Duration: {minutes:.2f} minutes")
        print(f"Parking Fee: â‚¹{fee:.2f}")
        print("===========================\n")

        del self.active_vehicles[number]
        return True
    
p1 = ParkingLot(3)

b1 = Bike("BIKE-123")
c1 = Car("CAR-456")
t1 = Truck("TRUCK-789")

p1.vehicle_entry(b1)
time.sleep(2)

p1.vehicle_entry(c1)
time.sleep(2)

p1.vehicle_exit("BIKE-123") 

        
p1.vehicle_entry(t1)
time.sleep(2)

p1.vehicle_exit("CAR-456")
p1.vehicle_exit("TRUCK-789")
```

```python
from abc import ABC , abstractmethod

class Resturant(ABC):
    def __init__(self , name):
        self.name = name 
        self.__orders = []
    
    def add_orders(self , item , price):
        self.__orders.append((item , price))
        
    def get_total_amount(self):
        return sum(price for _ , price in self.__orders)
        
```
